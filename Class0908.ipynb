```python
import numpy as np
import matplotlib.pyplot as plt
from pytictoc import TicToc
import math
```

```python
x = np.linspace(0,1, 10)
y = 2*x +1

plt.figure(figsize = (5,10)) # 빈 창을 하나 생성, figsize(가로크기, 세로크기)
plt.subplot(2,1,1) # subplot(나눌 행의수, 나눌 열의 수, 몇번째 위치)
plt.plot(x,y)

plt.figure(figsize = (5,10)) 
plt.subplot(2,1,2) # subplot(나눌 행의수, 나눌 열의 수, 몇번째 위치)
plt.scatter(x,y) # 산포도, 점으로, scatter(가로축,세로축)
```

```python
theta = np.linspace(0,2*math.pi, 100)
x = np.cos(theta) # 라디안으로 넣어야함
y = np.sin(theta)

plt.figure()
plt.subplot(2,1,1)
plt.plot(theta, x) # 앞에 있는건 아래에 깔리는 그림
plt.plot(theta, y) # 자동으로 겹쳐서, 뒤에 있는걸 위에 그림

plt.subplot(2,1,2)
plt.plot(x,y)
```

```python
import numpy.matlib

x = np.array([[1,2],[3,4]])
#x = x.reshape(3,1)
print(x)

X = np.matlib.repmat(x, 1, 2) # repmat(대상, ~행만큼, ~열만큼 반복) 1이면 반복안됨 곱하기라고 생각하자 
print(X)
```

```python
# 2차원, 100개의 랜덤 샘플을 생성 = 샘픔 x
# 2차원의 고정된 점 하나를 정함 = 기준점
# 기준점으로부터, 샘플까지의 거리를 계산
# 계산한 거리를 기준으로, 일정 거리 이상이 되는 애들을 따로 표시
# np.where() 괄호 안 조건문이 True인 인덱스를 반환하라
# for 문

t = TicToc()

N = 1000
D = 2
X = np.random.rand(N,D) # 행벡터 기준

# 임의로 잡은 기준점
z = np.array([0.3, 0.4])

# 거리 구하기 1 = for문
t.tic()
dist1 = np.zeros(N) # 거리를 담아둘 변수
for n in range(0,N):
    dist1[n] = np.sqrt((X[n,0] - z[0])**2 + (X[n,1] - z[1])**2) # SQRT(((0.54-0.3)**2 + (0.60-0.4)**2))
t.toc()

# 거리 구하기 2 = for문 안쓰고 repmat + sum으로
t.tic()
Z = np.matlib.repmat(z.reshape(1,2),N,1) # 샘플이랑 동일한 shape으로 생성  (100,2)차원으로 변경
dist2 = np.sqrt(np.sum((X - Z)**2, axis=1)) # 행렬의 각 원소의 차의 제곱   
t.toc()

d_star = 0.2 # 기준 거리
id_long = np.where(dist1 >= d_star)
    
plt.figure()
plt.scatter(X[id_long, 0], X[id_long, 1], s=50, facecolor = 'cyan', edgecolor = 'w')
plt.scatter(X[:,0], X[:,1], s=10, facecolor = 'gray')
plt.scatter(z[0], z[1], marker='x', facecolor='red')
```
